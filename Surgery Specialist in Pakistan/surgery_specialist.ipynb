### Handling 'Ratings' Column

```python
print(df['RATINGS'].dtype)
```

```python
df['RATINGS'] = df['RATINGS'].str.replace(',', '').str.strip()
df['RATINGS'] = pd.to_numeric(df['RATINGS'], errors='coerce')
```

```python
df['RATINGS'].dtype
```

```python
df['RATINGS'].head()
```

```python
max_rating = df['RATINGS'].max()
min_rating = df['RATINGS'].min()

print("Maximum rating:", max_rating)
print("Minimum rating:", min_rating)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.boxplot(df['RATINGS'])
plt.xlabel('Ratings')
plt.title('Boxplot of Ratings')
plt.show()
```

```python
max_index = df['RATINGS'].idxmax()

print("Index of maximum rating:", max_index)
```

```python
df.loc[max_index]
```

### Handling 'Satisfaction' Column

```python
print(df['SATISFACTION'].dtype)
```

```python
df['SATISFACTION'].unique()
```

```python
# Drop rows with non-numeric values in the 'SATISFACTION' column i.e remove '-' 
df = df[df['SATISFACTION'].str.rstrip('%').str.isdigit()]
```

```python
df.loc[:, 'SATISFACTION'] = pd.to_numeric(df['SATISFACTION'].str.rstrip('%'))
```

```python
df['SATISFACTION'].head()
```

```python
df.shape
```

```python
max_satisfaction = df['SATISFACTION'].max()
min_satisfaction = df['SATISFACTION'].min()

print("Maximum Satisfaction:", max_satisfaction)
print("Minimum Satisfaction:", min_satisfaction)
```

### No of Surgical Specialist with respect to Surgery Type in Pakistan

### RATINGS vs SATISFACTION

### Top 10 Ratings with Qualification 

```python
unique_doctors_df = df.drop_duplicates(subset='Name')
unique_doctors_df = unique_doctors_df[~unique_doctors_df['QUALIFICATIONS'].str.contains('-')]
top_ratings_with_qualifications = unique_doctors_df.nlargest(10, 'RATINGS')

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.barh(top_ratings_with_qualifications['QUALIFICATIONS'], top_ratings_with_qualifications['RATINGS'], color='skyblue')
plt.xlabel('Ratings')
plt.ylabel('Qualifications')
plt.title('Top 10 Ratings with Qualifications')
plt.tight_layout()
plt.show()
```

### Top-Rated Doctor for each unique Surgery Type:

```python
top_doctors = df.loc[df.groupby('SURGERY TYPE')['RATINGS'].idxmax(), ['SURGERY TYPE', 'Name', 'RATINGS', 'SATISFACTION']]

# Sort the DataFrame in descending order based on ratings
top_doctors_sorted = top_doctors.sort_values(by='RATINGS', ascending=False)

# Print the sorted DataFrame
print(top_doctors_sorted.to_string(index=False))
```
